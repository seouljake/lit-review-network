# import libraries

```python
import pandas as pd
```

# import data

```python
dat = pd.read_csv("../0 data_zotero/SK TW patronage.csv",encoding="UTF-8")
```

```python
# rename column names - remove spaces between words in each column label
newcolumns = {item:item.replace(" ","") for item in dat.columns}
dat = dat.rename(columns = newcolumns)
```

```python
dat
```

# Author - Clean & Disambiguate

```python
authorlist1 = dat[dat.Author.isnull() == False].Author.str.split(";").tolist()
```

```python
#authorlist
```

```python
# combine all authors into a single list
authorlist2 = []
for x in authorlist1:
    authorlist2+=x
```

```python
# remove whitespace from author names and save as a list, using a list comprehension
authorlist2 = [item.strip() for item in authorlist2]
```

```python
# remove duplicate names and sort alphabetically
length_with_duplicates = len(authorlist2)
print("With Duplicates:",length_with_duplicates)

authorlist2 = list(set(authorlist2))
authorlist2.sort()
length_without_duplicates = len(authorlist2)

print("Without Duplicates:",length_without_duplicates)

print("Duplicates:",length_with_duplicates-length_without_duplicates)
```

```python
# look over the list for problems (e.g. duplicates with slighly different spellings, Jr, institutes which are not authors)
authorlist2
```

```python
# export cleaned author names to .csv file
authorfile = open("../2 data cleaning_clean nodes in zotero/authors_disambiguated.csv",'w',encoding="utf-8")
for item in authorlist2:
    authorfile.write(item+"\n")
authorfile.close()
```

# Publication Titles of Journal Articles - Clean & Disambiguate

```python
dat.ItemType.unique()
```

```python
publist1 = dat[(dat.PublicationTitle.isnull() == False) & (dat.ItemType == "journalArticle")].PublicationTitle.str.strip()
```

```python
# remove 'The ' from publication title
publist1 = [item.replace("The ","") for item in publist1]
publist1
```

```python
# remove duplicate names and sort alphabetically
length_with_duplicates = len(publist1)
print("With Duplicates:",length_with_duplicates)

publist2 = list(set(publist1))
publist2.sort()
length_without_duplicates = len(publist2)

print("Without Duplicates:",length_without_duplicates)

print("Duplicates:",length_with_duplicates-length_without_duplicates)
```

```python
publist2
```

```python
# export cleaned publication names to .csv file
pubfile = open("../2 data cleaning_clean nodes in zotero/publication titles_disambiguated.csv",'w',encoding="utf-8")
for item in publist2:
    pubfile.write(item+"\n")
pubfile.close()
```
